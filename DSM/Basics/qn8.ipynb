```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 3, 5, 7, 9]

plt.plot(x1, y1, label='Line 1', marker='o')
plt.plot(x2, y2, label='Line 2', marker='s')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines on Same Plot')

plt.show()
```
